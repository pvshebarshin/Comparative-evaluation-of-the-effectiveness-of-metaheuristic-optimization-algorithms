```python
# В качестве метрик считаем
# 1) Среднее стандартное отклонение от оптимума
# 2) Среднее значение оптимума
# 3) Время выполнения алгоритма

file_name = "result.csv"
fields = ['Name_Evr', 'Name_Fun', 'Time', 'Epoch', 'Best_Fit', 'Std', 'Optimum', 'Parameters']


def make_evaluation(name, epoch, model, problem, optimum):
    # if len(optimums) != len(xy_opts):
    #     raise IndexError("Incorrect array dimension (len(optimums) != len(xy_opts))")
    exists = not os.path.exists(file_name)
    with open(file_name, 'a') as csvfile:
        csvwriter = csv.writer(csvfile)
        if exists:
            csvwriter.writerow(fields)
        start_time = time.time()
        best_position, best_fitness = model.solve(problem)
        end_time = time.time()
        # print(f"Best solution: {best_position}, Best fitness: {best_fitness}")
        row = [
            name,
            problem['fit_func'].__name__,
            end_time - start_time,
            epoch,
            best_fitness,
            fabs(optimum - best_fitness),
            optimum,
            'NO'
        ]
        csvwriter.writerow(row)
        csvfile.close()
```
