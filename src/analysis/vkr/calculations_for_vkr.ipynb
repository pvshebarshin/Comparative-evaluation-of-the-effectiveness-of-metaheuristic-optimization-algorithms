```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataset = pd.read_csv('../result.csv')
dataset.shift()
```

```python
IWO = dataset[(dataset['Name_Evr'] == 'IWO')]
HHO = dataset[(dataset['Name_Evr'] == 'HHO')]
ABC = dataset[(dataset['Name_Evr'] == 'ABC')]
ACO = dataset[(dataset['Name_Evr'] == 'ACO')]
BBO = dataset[(dataset['Name_Evr'] == 'BBO')]
CA = dataset[(dataset['Name_Evr'] == 'CA')]
DE = dataset[(dataset['Name_Evr'] == 'DE')]
FBIO = dataset[(dataset['Name_Evr'] == 'FBIO')]
GBO = dataset[(dataset['Name_Evr'] == 'GBO')]
PSO = dataset[(dataset['Name_Evr'] == 'PSO')]
```

```python
IWO_TE = IWO[['Time', 'Epoch']].groupby(['Epoch']).mean()
ACO_TE = ACO[['Time', 'Epoch']].groupby(['Epoch']).mean()
HHO_TE = HHO[['Time', 'Epoch']].groupby(['Epoch']).mean()
ABC_TE = ABC[['Time', 'Epoch']].groupby(['Epoch']).mean()
BBO_TE = BBO[['Time', 'Epoch']].groupby(['Epoch']).mean()
CA_TE = CA[['Time', 'Epoch']].groupby(['Epoch']).mean()
DE_TE = DE[['Time', 'Epoch']].groupby(['Epoch']).mean()
FBIO_TE = FBIO[['Time', 'Epoch']].groupby(['Epoch']).mean()
GBO_TE = GBO[['Time', 'Epoch']].groupby(['Epoch']).mean()
PSO_TE = PSO[['Time', 'Epoch']].groupby(['Epoch']).mean()
```

```python
plt.figure(figsize = [15,10])
plt.grid(True)

plt.plot(IWO_TE['Time'], label = 'IWO')
plt.plot(HHO_TE['Time'], label = 'HHO')
plt.plot(ABC_TE['Time'], label = 'ABC')
plt.plot(ACO_TE['Time'], label = 'ACO')
plt.plot(BBO_TE['Time'], label = 'BBO')
plt.plot(CA_TE['Time'], label = 'CA')
plt.plot(DE_TE['Time'], label = 'DE')
plt.plot(FBIO_TE['Time'], label = 'FBIO')
plt.plot(GBO_TE['Time'], label = 'GBO')
plt.plot(PSO_TE['Time'], label = 'PSO')

plt.ylabel('Time in seconds')
plt.xlabel('Epoch')
plt.title('Algorithms time')
plt.legend(loc=2, prop={'size': 20})
plt.show()
```

```python
import docx

df = dataset[dataset['Epoch'] == 200][dataset['Name_Evr'] != 'GA'][['Name_Evr', 'Time', 'Best_Fit', 'Std', 'Optimum']]\
.groupby(['Name_Evr']).mean()

doc = docx.Document()
# Initialise the table
t = doc.add_table(rows=(df.shape[0] + 1), cols=df.shape[1])
# Add the column headings
for j in range(df.shape[1]):
    t.cell(0, j).text = df.columns[j]
# Add the body of the data frame
for i in range(df.shape[0]):
    for j in range(df.shape[1]):
        cell = df.iat[i, j]
        t.cell(i + 1, j).text = str(cell)
# Save the Word doc
doc.save('table 2.docx')
```
