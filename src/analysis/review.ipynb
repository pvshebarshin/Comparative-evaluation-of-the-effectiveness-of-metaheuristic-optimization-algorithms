```python
x1 = np.linspace(-100, 100)
x2 = np.linspace(-100, 100)
X1, X2 = np.meshgrid(x1, x2)
plt.contour(X1, X2, easom_function([X1, X2]))
```

```python
def plotter(E, A):
    fig = plt.figure()
    ax = fig.add_subplot(projection='3d')
    surf = ax.plot_surface(X1, X2, easom_function([X1, X2]), \
        rstride=1, cstride=1, cmap=cm.jet, \
        edgecolor='darkred', linewidth=0.1)

    ax.set_xlabel('x1', fontsize=10)
    ax.set_ylabel('x2', fontsize=10)
    ax.set_zlabel('f(x1, x2)', fontsize=10)
    ax.set_title('Easom Function')
    ax.tick_params(axis='both', which='major', labelsize=6)
    ax.view_init(elev=E, azim=A)

iplot = interactive(plotter, E = (-60, 120, 5), A = (-60, 20, 5))
iplot
```
